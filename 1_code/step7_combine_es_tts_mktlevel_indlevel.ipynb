## Input: 
- market condition: tts_mktcondidtions
- price: price_for_es_from_ 
- es panel from step 4 output: es_panel_step4

```python
tts_mkt_condition=pd.read_csv('../2_pipeline/tts_mktcondidtions'+parameters+'.csv')
tts_priceinfo=pd.read_csv('../2_pipeline/price_for_es_from_'+parameters+'.csv')
es_panel=pd.read_csv('../2_pipeline/es_panel_step4'+parameters+'.csv')
```

```python
tts_mkt_condition.head()
```

```python
tts_mkt_condition=tts_mkt_condition.drop(columns=['Unnamed: 0'])
tts_mkt_condition.head()
```

```python
tts_priceinfo.head()
```

```python
tts_priceinfo=tts_priceinfo.drop(columns=['Unnamed: 0'])
tts_priceinfo.head()
```

```python
es_panel.head()
```

```python
es_panel=es_panel.drop(columns=['Unnamed: 0','Unnamed: 0.1'])
es_panel.head()
```

```python
es_panel.year_month_count.describe()
```

## merge market condition on es_panel 

```python
tts_mkt_condition['year_month_count']=(tts_mkt_condition.year-2013)*12+tts_mkt_condition.month
```

```python
tts_mkt_condition.head()
```

```python
tts_mkt_condition.describe()
```

```python
es_panel_merge1=pd.merge(left=es_panel,right=tts_mkt_condition,how='left',on=['market','year_month_count'])
```

```python
es_panel_merge1.tail()
```

## merge price info on es_panel 
Note that price info is on yearly basis 

```python
tts_priceinfo.head()
```

```python
es_panel_merge2=pd.merge(left=es_panel_merge1,right=tts_priceinfo,how='left',on=['market','year','installer_id'])
```

```python
es_panel_merge2.describe()
```

```python
es_panel_merge2.tail()
```

```python
es_panel_merge2=es_panel_merge2.drop(columns=['Unnamed: 0','Unnamed: 0.1'])
```

## Output: 2_pipeline/panel_step7+parameters.csv

```python
es_panel_merge2.to_csv('../2_pipeline/panel_step7'+parameters+'.csv')
```
